### Widgets

|        Widget          |         Description            |
| ---------------------- |  ----------------------------- |
| **Label**              | Just a label, not interective  |

```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()
root.title("Widgets Demo")

widgets = [
    tk.Label,
    tk.Checkbutton,
    ttk.Combobox,
    tk.Entry,
    tk.Button,
    tk.Radiobutton,
    tk.Scale,
    tk.Spinbox,
]
for widget in widgets:
    try:
        widget = widget(root, text=widget.__name__)
    except tk.TclError:
        widget = widget(root)
    widget.pack(padx=5, pady=5, fill="x")

root.mainloop()
```

### Button

It represents a clickable button that can trigger an action when pressed. 
You can customize the button by adjusting its width, height, font, bg (background color), and fg (foreground/text color)

```python
import tkinter as tk

root = tk.Tk()
root.title("TKinter Button Example")
root.geometry("250x150")

def on_click():
    label.config(text="Botón Pulsado")
    root.destroy()               # Closes the main Window, once the button is clicked

button = tk.Button(
    root,
    text="Acción",
    command=on_click,
)
button.pack(padx=5, pady=5)

# A helper label to show the result of the click
label = tk.Label(root, text="Esperando ser pulsado...")
label.pack(padx=5, pady=5)

root.mainloop()
```

### Checkbutton

it presents a checkable box to the user. 
You can programmatically set a checkbox state using *select()* or *deselect()*. You can access the state using a *tk.IntVar()* variable, which holds the checkbox's state: *1* for checked and *0* for unchecked.

```python
import tkinter as tk

rootW=tk.Tk()
rootW.title("TKinter Checkbutton")
rootW.geometry("250x90")

def show_state():
    checked = "Clica me" if var.get() else "Unchecked"
    checkbox.config(text=f"Clicame! ({checked})")

var = tk.IntVar()
checkbox = tk.Checkbutton (
    rootW,
    text="Clika me! (Cliklado)",
    variable=var
)
checkbox.select()
checkbox.config(command=show_state)
checkbox.pack(padx=5, pady=10)

rootW.mainloop()
```

### Combobox

The Combobox widget is a *drop-down list*, closed by default with an arrow to open it. 
You can select a single item from the list, and the currently selected item is shown as a label on the widget.
The combo box is suited to selecting a choice from a long list of options. 

You can add items to a Combobox by passing a list of strings to its values argument. The items will be added in the order we provide them:

```python
import tkinter as tk
from tkinter import ttk

rootW = tk.Tk()
rootW.title("TKinter Combobox")
rootW.geometry('250x90')

def selection_changed(event):
    label.config(text=f"{event.widget.get()} seleccionado!")

combobox = ttk.Combobox(rootW, values=["Primero", "Segundo", "Tercero"], state='normal')
combobox.set("primero")
combobox.pack(padx=5, pady=5, fill="x")

label = tk.Label(rootW, text=f"{combobox.get()} seleccionado!")
label.pack(pady=5)

combobox.bind('<<ComboboxSelected>>', selection_changed)

rootW.mainloop()
```

The **bind()** function connects the **ComboboxSelected** event to the **selection_changed()** function. This event is triggered when the currently selected item changes. The function updates the text of the helper label to reflect the selected item.


### Listbox
The Listbox widget is similar to Combobox, except that its options are presented as a scrollable list of items. It also supports the selection of multiple items at once. The Listbox class offers a <<ListboxSelect>> event that sends the selected item's index:

In this example, you create a list box and populate it through a for loop and the insert() function. Then, you connect the "<<ListboxSelect>>" event with the selection_changed() function. The helper label at the bottom of the window shows the selected item.

```python
import tkinter as tk

root = tk.Tk()
root.title("Tkinter Listbox")

def selection_changed(event):
    selection = event.widget.curselection()
    if selection:
        index = selection[0]
        label.config(text=f" has seleccionado: {event.widget.get(index)}!")
        event.widget.get(index)

listbox = tk.Listbox(root)
for item in ["el Primero", "el Segundo", "el Tercero", "el Cuarto"]:
    listbox.insert(tk.END, item)
listbox.bind("<<ListboxSelect>>", selection_changed)
listbox.pack(padx=5, pady=5, fill="both", expand=True)

# A helper label to show the selected value
label = tk.Label(root, text="Por favor selecciona")
label.pack(padx=5, pady=5, fill="x")

root.mainloop()

```

### Entry

The Entry widget is a simple single-line text editing box, into which users can type input. 

These are used for form fields, or settings where there is no restricted list of valid inputs. 

In this example, you create a text entry using the Entry widget. 
*     Then, you insert a placeholder text using the insert() function.
*     The Entry widget allows you to handle various text events, including when the Return key is pressed.
*     Binding this even to return_pressed() ensures that when you press Enter, the helper label displays the text that you type into the entry.

The **Entry class** has several useful features. They even allow you to perform input validation using custom validation functions and regular expressions.

```python
import tkinter as tk

rootW = tk.Tk()
rootW.title("Tkinter Entry")

def return_pressed(event):
    label.config(text=event.widget.get())

entry = tk.Entry(rootW)
entry.insert(0, "Enter your text")
entry.bind("<Return>", return_pressed)
entry.pack(padx=5, pady=5, fill="x")

# A helper label to show the selected value
label = tk.Label(rootW, text="Entry demo!")
label.pack(padx=5, pady=5, fill="x")

rootW.mainloop()
```

### Spinbox

The Spinbox widget provides an input box for numerical values. It has arrows to increase and decrease the value. It supports integers natively.

The code bellow, shows the various features available for the Spinbox widget. You can set the interval extremes with the from_ and to arguments. In practice, you'll often use the textvariable option to control the spin box value.

```python
import tkinter as tk

root = tk.Tk()
root.title("Tkinter Spinbox")
root.geometry("200x80")

spinbox_var = tk.StringVar(value="0")
spinbox = tk.Spinbox(
    root,
    from_=-10,
    to=10,
    textvariable=spinbox_var,
)
spinbox.pack(padx=5, pady=5, fill="x")

# A helper label to show the selected value
label = tk.Label(root, textvariable=spinbox_var)
label.pack(padx=5, pady=5, fill="x")

root.mainloop()
```

### Scale

The Scale widget provides a slide-bar widget that works much like a Spinbox. Rather than displaying the current value numerically, it displays the position of the slider handle along the length of the widget.

This widget is often useful when we need to adjust between two extremes but where absolute accuracy is not required. The most common use of this type of widget is for volume controls in multimedia apps:

```python
import tkinter as tk

root = tk.Tk()
root.title("Tkinter Scale")
root.geometry("180x190")

def value_changed(event):
    label.config(text=event.widget.get())

# Scale for Temperature
scaleT = tk.Scale(root, from_=10, to=-10, orient="vertical")
scaleT.bind("<Motion>", value_changed)
scaleT.pack(padx=5, pady=5, fill="x")

# Scale for Pressure
scaleP = tk.Scale(root, from_=0, to=320, orient="horizontal")
scaleP.bind("<Motion>", value_changed)
scaleP.pack(padx=5, pady=5, fill="x")

# A helper label to show the selected value
label = tk.Label(root, text="0")
label.pack(padx=5, pady=5, fill="x")

root.mainloop()
```

Tkinter offers three primary layout managers for positioning widgets, including Frames, within a window: pack(), grid(), and place(). Each has its own strengths:

* **pack()**: Organizes widgets in blocks before placing them in the parent widget. It's simple for basic layouts (top, bottom, left, right).
* **grid()**: Arranges widgets in a table-like structure of rows and columns. It's excellent for more complex, aligned layouts.
* **place()**: Allows you to position widgets at absolute coordinates or relative to the parent's size. It offers the most precise control but can be harder to manage for responsive designs.


```python
import tkinter as tk

def create_window_with_frame_pack():
    window = tk.Tk()
    window.title("Tkinter Frame with pack()")
    window.geometry("300x200")

    # Create a Frame
    my_frame = tk.Frame(window, bg="lightblue", bd=2, relief="solid")

    # Position the Frame using pack()
    # fill='both' expands the frame to fill available space
    # expand=True allows the frame to expand when the window is resized
    my_frame.pack(pady=20, padx=20, fill='both', expand=True)

    label = tk.Label(my_frame, text="Frame positioned with pack()", bg="lightblue")
    label.pack(pady=10)

    window.mainloop()

def create_window_with_frame_grid():
    window = tk.Tk()
    window.title("Tkinter Frame with grid()")
    window.geometry("300x200")

    # Configure the grid of the main window to make it resizable
    window.grid_rowconfigure(0, weight=1)
    window.grid_columnconfigure(0, weight=1)

    # Create a Frame
    my_frame = tk.Frame(window, bg="lightgreen", bd=2, relief="solid")

    # Position the Frame using grid()
    # row and column specify its position
    # sticky='nsew' makes the frame expand to fill the cell (North, South, East, West)
    my_frame.grid(row=0, column=0, pady=20, padx=20, sticky='nsew')

    label = tk.Label(my_frame, text="Frame positioned with grid()", bg="lightgreen")
    label.pack(pady=10)

    window.mainloop()

def create_window_with_frame_place():
    window = tk.Tk()
    window.title("Tkinter Frame with place()")
    window.geometry("300x200")

    # Create a Frame
    my_frame = tk.Frame(window, bg="lightcoral", bd=2, relief="solid")

    # Position the Frame using place()
    # relx and rely are relative x and y coordinates (0.0 to 1.0)
    # anchor='center' centers the frame at the specified relx/rely
    # relwidth and relheight are relative width and height
    my_frame.place(relx=0.5, rely=0.5, relwidth=0.7, relheight=0.5, anchor='center')

    label = tk.Label(my_frame, text="Frame positioned with place()", bg="lightcoral")
    label.pack(pady=10)

    window.mainloop()

# --- Run the examples one by one ---
# Uncomment the function call you want to test

# create_window_with_frame_pack()
# create_window_with_frame_grid()
create_window_with_frame_place()
```
